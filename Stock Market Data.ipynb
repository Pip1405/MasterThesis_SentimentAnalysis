# ENERGY STOCK PRICES

```python
import pandas as pd
import matplotlib.pyplot as plt

# Excel-Datei einlesen
data = pd.read_excel('historical_data/Energy.xlsx')

# Datenbereich ausw채hlen (ohne die erste Zeile mit den Spalten체berschriften) und nur die Spalten A bis I
data = data.iloc[1:, :9]

# Spaltennamen f체r die Titel der Aktien extrahieren
stock_names = data.columns[1:]

# Aktienklarnamen
stock_full_names = {
    'XOM.N': 'Exxon Mobil Corp',
    'CVX': 'Chevron Corp',
    'COP.N': 'Conocophillips',
    'SLB.N': 'Schlumberger NV',
    'EOG.N': 'EOG Resources Inc',
    'EPD.N': 'Enterprise Products Partners LP',
    'OXY.N': 'Occidental Petroleum Corp',
    'MPC.N': 'Marathon Petroleum Corp'
}

# Monatliche Rendite berechnen
monthly_returns = data.groupby(pd.to_datetime(data['Date']).dt.to_period('M')).last()[stock_names].pct_change()

# Durchschnittliche monatliche Rendite berechnen
average_returns = monthly_returns.mean() * 100

# Volatilit채t (Standardabweichung) berechnen
volatility = monthly_returns.std() * 100

# Gesamtrendite 체ber den gesamten Zeitraum berechnen
total_returns = ((data.iloc[-1, 1:] / data.iloc[0, 1:]) - 1) * 100

# Durchschnittliche Jahresrendite berechnen
yearly_returns = ((1 + monthly_returns.mean()) ** 12 - 1) * 100

# Ergebnisse in eine Tabelle umwandeln
results = pd.DataFrame({
    'Aktie': stock_names,
    'Aktienklarname': [stock_full_names[stock] for stock in stock_names],
    'Durchschnittliche monatliche Rendite (%)': average_returns,
    'Volatilit채t (Standardabweichung) (%)': volatility,
    'Gesamtrendite 체ber den gesamten Zeitraum (%)': total_returns,
    'Durchschnittliche Jahresrendite (%)': yearly_returns
})

# Tabelle anzeigen
display(results)

# Aktienverlauf plotten
fig, ax = plt.subplots(figsize=(12, 6))  # Gr철e der Abbildung anpassen
for stock_name in stock_names:
    ax.plot(data['Date'], data[stock_name], label=stock_name)

# Plot anpassen
ax.set_xlabel('Date')
ax.set_ylabel('Close Price')
ax.set_title('Stock Prices - Energy')
ax.legend(loc='upper left')  # Legendenposition 채ndern

# Diagramm anzeigen
plt.tight_layout()  # Platz f체r Labels und Titel optimieren
plt.show()
```

# RETURN CALCULATION FOR CSV

```python
import pandas as pd

# Daten aus der Excel-Datei einlesen und invertieren
data = pd.read_excel("historical_data/Technology.xlsx").iloc[::-1]

# Spaltennamen der Aktienkurse extrahieren
stock_columns = data.columns[1:9]

# Neue DataFrame erstellen, um die Renditen zu speichern
returns_data = pd.DataFrame(columns=["Date"] + list(stock_columns))

# Schleife 체ber die Datenpunkte in umgekehrter Reihenfolge
for i in range(len(data)-1, 0, -1):
    date = data.loc[i, "Date"]
    
    # Umwandlung des Datumsformats in "YYYY-MM-DD"
    date = pd.to_datetime(date, format="%d/%m/%Y").strftime("%Y-%m-%d")
    
    # Schleife 체ber die Aktienkurse
    for stock in stock_columns:
        # Konvertierung der Werte in numerischen Typ (falls erforderlich)
        prev_value = pd.to_numeric(data.loc[i, stock], errors="coerce")
        curr_value = pd.to_numeric(data.loc[i-1, stock], errors="coerce")
        
        if pd.notnull(prev_value) and pd.notnull(curr_value):
            return_value = (curr_value - prev_value) / prev_value
            
            # Renditen in DataFrame speichern
            returns_data.loc[i-1, stock] = return_value

# Erstellen einer neuen Spalte mit korrektem Datum
returns_data["Date"] = pd.to_datetime(data.iloc[1:, 0], format="%d/%m/%Y").dt.strftime("%Y-%m-%d")

# CSV-Datei speichern
returns_data.to_csv("historical_data/Technology_returns.csv", index=False)

# Ausgabe der Renditen
print(returns_data)
```

# SECTOR PORTFOLIO (BASE 100) FULL 

```python
import pandas as pd
import matplotlib.pyplot as plt

def calculate_portfolio_performance(table_name):
    # Daten einlesen
    weights_data = pd.read_excel("historical_data/{}.xlsx".format(table_name), sheet_name="Table Data", header=1)
    returns_data = pd.read_csv("historical_data/{}_returns.csv".format(table_name), header=None)

    # Portfolio-Gewichte extrahieren und invertieren
    portfolio_weights = weights_data.iloc[:, 18:26].values[:-1]

    # Renditen extrahieren und invertieren
    returns = returns_data.iloc[1:, 1:].values.astype(float)

    # Portfolio-Renditen berechnen
    portfolio_returns = (portfolio_weights * returns).sum(axis=1)

    return_0 = portfolio_weights[0] * returns[0]
    print("portfolio_weights",portfolio_weights[0] )
    print("returns:", returns[0])
    print("results:",return_0)
    print( "final:",return_0.sum())

    # T채gliche Rendite des Portfolios
    portfolio_daily_return = portfolio_returns[0]

    # Portfolioverlauf simulieren
    portfolio_values = [100]  # Anfangswert von 100
    for i in range(1, len(portfolio_returns)):
        portfolio_value = portfolio_values[i-1] * (1 + portfolio_returns[i])
        portfolio_values.append(portfolio_value)

    # Durchschnittliche Rendite berechnen
    average_return = portfolio_returns.mean()

    # Datumsspalte invertieren und hinzuf체gen
    dates = returns_data.iloc[1:, 0].values
    portfolio_returns_data = pd.DataFrame({"Date": dates, "Portfolio Return": portfolio_returns, "Portfolio Value": portfolio_values})

    # Portfolioverlauf grafisch darstellen
    fig, ax = plt.subplots()
    ax.plot(portfolio_returns_data["Date"], portfolio_returns_data["Portfolio Value"])
    ax.set_xlabel("Date")
    ax.set_ylabel("Portfolio Value")
    ax.set_title("{} Portfolio Performance".format(table_name))
    
    # Anpassung der x-Achsenbeschriftung
    num_ticks = 10  # Anzahl der angezeigten Werte
    x_ticks = range(0, len(portfolio_returns_data), len(portfolio_returns_data) // num_ticks)
    x_labels = [date.split("-")[0] for date in portfolio_returns_data["Date"]][::len(portfolio_returns_data) // num_ticks]
    ax.set_xticks(x_ticks)
    ax.set_xticklabels(x_labels, rotation=45)
    
    ax.grid(True)
    plt.show()

    # Durchschnittliche Rendite ausgeben
    print("Durchschnittliche t채gliche Rendite f체r {}: {}".format(table_name, average_return))

    # Portfolio-Daten in CSV-Datei schreiben
    portfolio_returns_data.to_csv("historical_data/{}_portfolio.csv".format(table_name), index=False)


# Portfolio Performance f체r Technology
calculate_portfolio_performance("Technology")

# Portfolio Performance f체r Financials
calculate_portfolio_performance("Financials")

# Portfolio Performance f체r Energy
calculate_portfolio_performance("Energy")
```

# MARKET CAP VISUALIZATION

```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten einlesen
market_cap_data = pd.read_excel("historical_data/Technology.xlsx", sheet_name="Table Data", header=2, usecols="A:Q", skiprows=2)

# Datumsspalte extrahieren und umkehren
dates = market_cap_data.iloc[:, 0].values[::-1]

# Marktkapitalisierungsspalten extrahieren und umkehren
market_caps = market_cap_data.iloc[:, 9:].values[::-1]

# DataFrame f체r den Verlauf der Marktkapitalisierung erstellen
market_cap_df = pd.DataFrame(data=market_caps, columns=market_cap_data.columns[9:], index=dates)

# Marktkapitalisierung Verlauf grafisch darstellen
plt.figure(figsize=(10, 6))
for column in market_cap_df.columns:
    plt.plot(market_cap_df.index, market_cap_df[column], label=column)

plt.xlabel("Date")
plt.ylabel("Market Cap")
plt.title("Energy Market Cap")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

# SECTOR PORTFOLIO (BASE 100) 2022 

### RETURNS

```python
import pandas as pd


def calculate_portfolio_performance(table_name):
    # Daten einlesen
    data = pd.read_csv("historical_data/{}_portfolio.csv".format(table_name))

    # Berechnung des nominalen Portfoliowerts ausgehend von 100
    data["Portfolio Value"] = (1 + data["Portfolio Return"]).cumprod() * 100

    return data

# Portfolio Performance f체r Technology
technology_data = calculate_portfolio_performance("Technology")
technology_data.rename(columns={"Portfolio Return": "Technology Return", "Portfolio Value": "Technology Portfolio"}, inplace=True)

# Portfolio Performance f체r Financials
financials_data = calculate_portfolio_performance("Financials")
financials_data.rename(columns={"Portfolio Return": "Financials Return", "Portfolio Value": "Financials Portfolio"}, inplace=True)

# Portfolio Performance f체r Energy
energy_data = calculate_portfolio_performance("Energy")
energy_data.rename(columns={"Portfolio Return": "Energy Return", "Portfolio Value": "Energy Portfolio"}, inplace=True)

# Zusammenf체hren der Daten f체r alle Portfolios
portfolio_results = pd.merge(technology_data[["Date", "Technology Return"]],
                             financials_data[["Date", "Financials Return"]],
                             on="Date")
portfolio_results = pd.merge(portfolio_results,
                             energy_data[["Date", "Energy Return"]],
                             on="Date")

# Filtern nach Jahr 2022
portfolio_results['Date'] = pd.to_datetime(portfolio_results['Date'])
mask = (portfolio_results['Date'].dt.year == 2022)
portfolio_results = portfolio_results.loc[mask]

# Portfolio-Daten in CSV-Datei schreiben
portfolio_results.to_csv("Portfolio_Returns_Baseline.csv", index=False)
```
